```python
import pandas

# read the xlsx file "plot.xlsx" into a pandas dataframe
df = pandas.read_excel("plot.xlsx")

# get the colomn name of the dataframe
name = df.columns
```

```python
# get the data of each colomn
erasing_ratio = df['erasing_ratio']
high_cam_random_vit = df['high_cam_random_vit']
high_cam_random_resnet18 = df['high_cam_random_resnet18']
low_cam_random_vit = df['low_cam_random_vit']
low_cam_random_resnet18 = df['low_cam_random_resnet18']
global_random_vit = df['global_random_vit']
global_random_vit_resnet18 = df['global_random_resnet18']
high_cam_vit = df['high_cam_vit']
high_cam_resnet18 = df['high_cam_resnet18']
low_cam_vit = df['low_cam_vit']
low_cam_resnet18 = df['low_cam_resnet18']
```

```python
# print the data of each colomn
print(erasing_ratio)
print(high_cam_random_vit)
print(high_cam_random_resnet18)
print(low_cam_random_vit)
print(low_cam_random_resnet18)
print(global_random_vit)
print(global_random_vit_resnet18)
print(high_cam_vit)
print(high_cam_resnet18)
print(low_cam_vit)
print(low_cam_resnet18)


# plot the figure of vit and resnet18 with different erasing ratio, and the VIT model uses a solid line and the RESNet 18 uses a dashed line
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(erasing_ratio, high_cam_random_vit, label='high_cam_random_vit', marker='o')
plt.plot(erasing_ratio, high_cam_random_resnet18, label='high_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_random_vit, label='low_cam_random_vit', marker='o')
plt.plot(erasing_ratio, low_cam_random_resnet18, label='low_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, global_random_vit, label='global_random_vit', marker='o')
plt.plot(erasing_ratio, global_random_vit_resnet18, label='global_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, high_cam_vit, label='high_cam_vit', marker='o')
plt.plot(erasing_ratio, high_cam_resnet18, label='high_cam_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_vit, label='low_cam_vit', marker='o')
plt.plot(erasing_ratio, low_cam_resnet18, label='low_cam_resnet18', linestyle='--', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()


# just plot the figure of vit with different ratio
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(erasing_ratio, high_cam_random_vit, label='high_cam_random_vit', marker='o')
plt.plot(erasing_ratio, low_cam_random_vit, label='low_cam_random_vit', marker='o')
plt.plot(erasing_ratio, global_random_vit, label='global_random_vit', marker='o')
plt.plot(erasing_ratio, high_cam_vit, label='high_cam_vit', marker='o')
plt.plot(erasing_ratio, low_cam_vit, label='low_cam_vit', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()

# just plot the figure of resnet18 with different ratio
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(erasing_ratio, high_cam_random_resnet18, label='high_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_random_resnet18, label='low_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, global_random_vit_resnet18, label='global_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, high_cam_resnet18, label='high_cam_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_resnet18, label='low_cam_resnet18', linestyle='--', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()
```

```python
import pandas
# read the file plot.xlsx "vit", "resnet18", "wide_resnet"
df1 = pandas.read_excel("plot.xlsx", sheet_name='vit')
df2 = pandas.read_excel("plot.xlsx", sheet_name='resnet18')
df3 = pandas.read_excel("plot.xlsx", sheet_name='wide_resnet50')


# get the columns of df1, df2, df3, and read the data in dataframe 
df1.columns
df2.columns
df3.columns
```

```python
# plot the figure of three model, and x-axis is erasing_ratio, the y-axis is accuracy, there are three lien in figure, one is vit, one is resnet18, one is wide_resnet50
# the columns of one model is
# erasing_ratio', 'global_random_erasing', 'front_random_erasing',
    #    'back_random_erasing', 'front_erasing', 'back_erasing'],
    #   dtype='object'
vit_global_random_erasing = df1['global_random_erasing']
resnet18_global_random_erasing = df2['global_random_erasing']
wide_resnet50_global_random_erasing = df3['global_random_erasing']

vit_front_random_erasing = df1['front_random_erasing']
resnet18_front_random_erasing = df2['front_random_erasing']
wide_resnet50_front_random_erasing = df3['front_random_erasing']

vit_back_random_erasing = df1['back_random_erasing']
resnet18_back_random_erasing = df2['back_random_erasing']
wide_resnet50_back_random_erasing = df3['back_random_erasing']

vit_front_erasing = df1['front_erasing']
resnet18_front_erasing = df2['front_erasing']
wide_resnet50_front_erasing = df3['front_erasing']

vit_back_erasing = df1['back_erasing']
resnet18_back_erasing = df2['back_erasing']
wide_resnet50_back_erasing = df3['back_erasing']

vit_erasing_ratio = df1['erasing_ratio']
resnet18_erasing_ratio = df2['erasing_ratio']
wide_resnet50_erasing_ratio = df3['erasing_ratio']

# plot three figure, and each include the result of three model in different erasing_ratio, and each model have five line include global_random_erasing, fornt_eraing and so on.
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
# set the stype
plt.style.use('seaborn')
# plot the first figure
plt.plot(vit_erasing_ratio, vit_global_random_erasing, label='vit_global_random_erasing', marker='o')
plt.plot(vit_erasing_ratio, vit_front_random_erasing, label='vit_front_random_erasing', marker='o')
plt.plot(vit_erasing_ratio, vit_back_random_erasing, label='vit_back_random_erasing', marker='o')
plt.plot(vit_erasing_ratio, vit_front_erasing, label='vit_front_erasing', marker='o')
plt.plot(vit_erasing_ratio, vit_back_erasing, label='vit_back_erasing', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.savefig('vit.svg', format='svg')
plt.show()
# plot the resnet18 figure
plt.plot(resnet18_erasing_ratio, resnet18_global_random_erasing, label='resnet18_global_random_erasing', marker='o')
plt.plot(resnet18_erasing_ratio, resnet18_front_random_erasing, label='resnet18_front_random_erasing', marker='o')
plt.plot(resnet18_erasing_ratio, resnet18_back_random_erasing, label='resnet18_back_random_erasing', marker='o')
plt.plot(resnet18_erasing_ratio, resnet18_front_erasing, label='resnet18_front_erasing', marker='o')
plt.plot(resnet18_erasing_ratio, resnet18_back_erasing, label='resnet18_back_erasing', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.savefig('resnet18.svg', format='svg')
plt.show()
# plot the wide_resnet50 figure
plt.plot(wide_resnet50_erasing_ratio, wide_resnet50_global_random_erasing, label='wide_resnet50_global_random_erasing', marker='o')
plt.plot(wide_resnet50_erasing_ratio, wide_resnet50_front_random_erasing, label='wide_resnet50_front_random_erasing', marker='o')
plt.plot(wide_resnet50_erasing_ratio, wide_resnet50_back_random_erasing, label='wide_resnet50_back_random_erasing', marker='o')
plt.plot(wide_resnet50_erasing_ratio, wide_resnet50_front_erasing, label='wide_resnet50_front_erasing', marker='o')
plt.plot(wide_resnet50_erasing_ratio, wide_resnet50_back_erasing, label='wide_resnet50_back_erasing', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.savefig('wide_resnet50.svg', format='svg')
plt.show()


```
