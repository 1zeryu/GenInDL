```python
import pandas

# read the xlsx file "plot.xlsx" into a pandas dataframe
df = pandas.read_excel("plot.xlsx")

# get the colomn name of the dataframe
name = df.columns
```

```python
name
```

```python
# get the data of each colomn
erasing_ratio = df['erasing_ratio']
high_cam_random_vit = df['high_cam_random_vit']
high_cam_random_resnet18 = df['high_cam_random_resnet18']
low_cam_random_vit = df['low_cam_random_vit']
low_cam_random_resnet18 = df['low_cam_random_resnet18']
global_random_vit = df['global_random_vit']
global_random_vit_resnet18 = df['global_random_resnet18']
high_cam_vit = df['high_cam_vit']
high_cam_resnet18 = df['high_cam_resnet18']
low_cam_vit = df['low_cam_vit']
low_cam_resnet18 = df['low_cam_resnet18']
```

```python
# print the data of each colomn
print(erasing_ratio)
print(high_cam_random_vit)
print(high_cam_random_resnet18)
print(low_cam_random_vit)
print(low_cam_random_resnet18)
print(global_random_vit)
print(global_random_vit_resnet18)
print(high_cam_vit)
print(high_cam_resnet18)
print(low_cam_vit)
print(low_cam_resnet18)


# plot the figure of vit and resnet18 with different erasing ratio, and the VIT model uses a solid line and the RESNet 18 uses a dashed line
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(erasing_ratio, high_cam_random_vit, label='high_cam_random_vit', marker='o')
plt.plot(erasing_ratio, high_cam_random_resnet18, label='high_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_random_vit, label='low_cam_random_vit', marker='o')
plt.plot(erasing_ratio, low_cam_random_resnet18, label='low_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, global_random_vit, label='global_random_vit', marker='o')
plt.plot(erasing_ratio, global_random_vit_resnet18, label='global_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, high_cam_vit, label='high_cam_vit', marker='o')
plt.plot(erasing_ratio, high_cam_resnet18, label='high_cam_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_vit, label='low_cam_vit', marker='o')
plt.plot(erasing_ratio, low_cam_resnet18, label='low_cam_resnet18', linestyle='--', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()


# just plot the figure of vit with different ratio
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(erasing_ratio, high_cam_random_vit, label='high_cam_random_vit', marker='o')
plt.plot(erasing_ratio, low_cam_random_vit, label='low_cam_random_vit', marker='o')
plt.plot(erasing_ratio, global_random_vit, label='global_random_vit', marker='o')
plt.plot(erasing_ratio, high_cam_vit, label='high_cam_vit', marker='o')
plt.plot(erasing_ratio, low_cam_vit, label='low_cam_vit', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()

# just plot the figure of resnet18 with different ratio
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(erasing_ratio, high_cam_random_resnet18, label='high_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_random_resnet18, label='low_cam_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, global_random_vit_resnet18, label='global_random_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, high_cam_resnet18, label='high_cam_resnet18', linestyle='--', marker='o')
plt.plot(erasing_ratio, low_cam_resnet18, label='low_cam_resnet18', linestyle='--', marker='o')
plt.xlabel('erasing_ratio')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()
```

```python
import pandas

# read the plot.xlsx file and get the second table
df2 = pandas.read_excel("plot.xlsx", sheet_name='Sheet2')

df2.columns
```

```python
# get the data of each colomn odf df2
space = df2['space']
high_vit = df2['high_vit']
high_resnet18 = df2['high_resnet18']
random_vit = df2['random_vit']
random_resnet18 = df2['random_resnet18']
low_vit = df2['low_vit']
low_resnet18 = df2['low_resnet18']

# plot the figure with x axis is eraisng_ratio and y axis is accuracy, you should plot the figure of vit and resnet18 with different erasing ratio, and the VIT model uses a solid line and the RESNet 18 uses a dashed line
# you can just use there color for each process high, random, low
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
# high vit and high resnet18 with same color
plt.plot(space, high_vit, label='high_vit', marker='o', color='blue')
plt.plot(space, high_resnet18, label='high_resnet18', linestyle='--', marker='o', color='blue')
# random vit and random resnet18 with same color
plt.plot(space, random_vit, label='random_vit', marker='o', color='red')
plt.plot(space, random_resnet18, label='random_resnet18', linestyle='--', marker='o', color='red')
# low vit and low resnet18 with same color
plt.plot(space, low_vit, label='low_vit', marker='o', color='green')
plt.plot(space, low_resnet18, label='low_resnet18', linestyle='--', marker='o', color='green')
plt.xlabel('space')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()
```
